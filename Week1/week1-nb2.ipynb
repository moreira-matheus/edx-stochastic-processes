```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(-3,3)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
def func(x, n):
    y = x**n
    return y
```

```python
x = np.logspace(-2,2)

for n in [1,2,3]:
    plt.plot(x, func(x,n), label='$x^{}$'.format(n))

plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.show()
```

```python
N = 100000

np.random.seed(0)
R = np.random.randn(N)

plt.hist(R, bins=100, density=True)
plt.show()
```

```python
NSTEP = 10000

plt.xlabel('num. of steps')
plt.ylabel('position')

for nseed in range(10):
    np.random.seed(nseed)
    step = np.random.choice([-1,1], NSTEP)
    position = np.cumsum(step)
    plt.plot(position)
```

***

```python
num_trajs = 1000
num_steps = 10000

avg_traj = np.zeros(num_steps)

fig, ax = plt.subplots(ncols=2, sharey=True)

for _ in range(num_trajs):
    steps = np.random.choice([-1, 1], num_steps)
    traj = np.cumsum(steps)

    ax[0].plot(traj, linewidth=0.5)

    avg_traj = avg_traj + traj

avg_traj = avg_traj/num_trajs

ax[1].plot(avg_traj, color='k')


ax[0].set_title('All trajectories ({})'.format(num_trajs))
ax[1].set_title('Average trajectory')

ax[0].axhline(y=0, xmin=0, xmax=num_steps, linestyle='dotted', color='gray')
ax[1].axhline(y=0, xmin=0, xmax=num_steps, linestyle='dotted', color='gray')

plt.show()
```
